```python

import yfinance as yf
bitcoin=yf.Ticker("BTC-USD")
bitcoin=bitcoin.history(period='max')
bitcoin
```

```python
bitcoin.index
```

Preprocessiong and Data Visualization

```python
bitcoin.plot.line(y='Close',use_index=True)
```

```python
bitcoin['Tmr']=bitcoin['Close'].shift(-1)
```

```python
bitcoin
```

```python
bitcoin['Target']=(bitcoin['Tmr']>bitcoin['Close']).astype(int)
```

```python
bitcoin
```

Training an Initial Model

```python
#im using radnomforestclassifier because it works on bunch of individual decision tress with randomized parameters
#its hard to overfit & also runs Quickly
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100,min_samples_split=100,random_state=1)
train=bitcoin.iloc[:-100]
test=bitcoin.iloc[-100:]
predictors=["Close","Volume","Open","High","Low"]
model.fit(train[predictors],train['Target'])
```

```python
from sklearn.metrics import precision_score
preds=model.predict(test[predictors])
```

```python
import pandas as pd
preds=pd.Series(preds,index=test.index)
preds
```

```python
precision_score(test['Target'],preds)
```

the above precision is not that good -- we need more good precision_score

```python
combined=pd.concat([test['Target'],preds],axis=1)
combined.plot()
```

Orange is 0 And Blue is 1

Building a Backtesting System

```python
#Defineing a Fun
def predict(train,test,predictors,mode):
  model.fit(train[predictors],train['Target'])
  preds=model.predict(test[predictors])
  preds=pd.Series(preds,index=test.index,name="Predictions")
  combined=pd.concat([test['Target'],preds],axis=1)
  return combined
```

```python
def backtest(data,model,predictors,start=2500,step=250):
  #here start is 10 years of data, ie we train 1st model with 10yrs of data then predict values for the 11th year and so on
  all_predictions=[]
  for i in range(start,data.shape[0],step):
    train=data.iloc[0:i].copy()
    test=data.iloc[i:(i+step)].copy()
    predictions=predict(train,test,predictors,model)
    all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions=backtest(bitcoin,model,predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions['Target'],predictions['Predictions'])
```
