Filling the missing spaces

```python
a=df['Age'].median()
df['Age'].fillna(a,inplace=True)
b=df['Embarked'].mode()[0]
df['Embarked'].fillna(b,inplace=True)
c=df['Fare'].mode()[0]
df['Fare'].fillna(c,inplace=True)
```

```python
df.isna().sum()
```

Converting Cateogry variables to dummy variable

(Help avoiding multicollinearity issues)

```python
df1=pd.get_dummies(df[['Sex','Embarked']],drop_first=True)
```

```python
df2=pd.concat([df,df1],axis=1)
df3=df2.drop(['Sex','Embarked'],axis=1)
```

```python
df3
```

```python
x=df3.drop(['Survived'],axis=1)
y=df3['Survived']
```

```python
x
```

```python
y
```

Spliting Dataset using train_test_split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

Standardizing the features in Dataset

(Standardization a process of rescaling mean =0 and deviation =1)


```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)

#x_train and x_test will be suitable for ML Algoritms like KNN,neural networks
```

Model Creation

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
#MODEL 1 KNeighborsClassifier
model1=KNeighborsClassifier(n_neighbors=7)
model1.fit(x_train,y_train)
y_pred=model1.predict(x_test)
print("KNearest Neighbor")
print("Comfussion Matrix")
print(confusion_matrix(y_test,y_pred))
print("Accuracy Score",accuracy_score(y_test,y_pred))
print("Classification Score")
print(classification_report(y_test,y_pred))
```

```python
#MODEL 2 Naive Bayes
model2=BernoulliNB()
model2.fit(x_train,y_train)
y_pred=model2.predict(x_test)
print("Naive Bayes")
print("Confussion Matrix")
print(confusion_matrix(y_test,y_pred))
print("Accuracy Score",accuracy_score(y_test,y_pred))
print("Classificaton Score")
print(classification_report(y_test,y_pred))
```

```python
#MODEL 3 SVC
model3=SVC()
model3.fit(x_train,y_train)
y_pred=model3.predict(x_test)
print("SVC")
print("Confussion Matrix")
print(confusion_matrix(y_test,y_pred))
print("Accuracy Score",accuracy_score(y_test,y_pred))
print("Classification Report")
print(classification_report(y_test,y_pred))
```

```python
#MODEL 4 DecisionTreeClassifier
model4=DecisionTreeClassifier(random_state=42)
model4.fit(x_train,y_train)
y_pred=model4.predict(x_test)
print("DecisionTreeClassifier")
print("Confussion Matrix")
print(confusion_matrix(y_test,y_pred))
print("Accuracy Score",accuracy_score(y_test,y_pred))
print("Classification Score")
print(classification_report(y_test,y_pred))
```

```python
#MODEL5 RandomForestClassifer
model5=RandomForestClassifier()
model5.fit(x_train,y_train)
y_pred=model5.predict(x_test)
print("RandomForestClassifer")
print("Confussion MAtrix")
print(confusion_matrix(y_test,y_pred))
print("Accuracy Score",accuracy_score(y_test,y_pred))
print("Classification Report")
print(classification_report(y_test,y_pred))
```
